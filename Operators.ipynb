```python
# function for arithmetic operators(+,-,*,/)

num1 = input ("num1 = ")
num2 = input ("num2 = ")

# sum of two numbers
print ("sum = ", float(num1) + float(num2))

# substraction of two numbers
print ("substraction = ", float(num1) - float(num2))

# multiplication of two numbers
print ("multiplication = ", float(num1) * float(num2))

# division of two numbers
print ("division = ", float(num1) / float(num2))
```

```python
# method for increment and decrement operators(++, --)

num = 15

print("Original number:", num)

# Increment
num += 1
print("After increment :", num)

# Decrement
num -= 1
print("After decrement :", num)
```

```python
# program to find the two numbers equal or not.

a = input("enter num1:")
b = input("enter num2:")

if a == b:
    print("both a and b are equal")
else:
    print("both a and b are not equal")
```

```python
# Program for relational operators (<,<==, >, >==)

a = 4
b = 2

print("a =", a)
print("b =", b)

print("a < b :", a < b)
print("a <== b:", a <= b)
print("a > b :", a > b)
print("a >== b:", a >= b)
print("a == b:", a == b)
print("a != b:", a != b)
```

```python
# Print the smaller and larger number

a = input("num1:")
b = input("num2:")

print("First number:", a)
print("Second number:", b)

# For printing the smaller number

if a < b:
    print("Smaller number is:", a)
else:
    print("Smaller number is:", b)

# For printing the larger number

if a > b:
    print("Larger number is:", a)
else:
    print("Larger number is:", b)
```
